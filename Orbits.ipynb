```python
%matplotlib notebook
import math 
import numpy as np 
import matplotlib.pyplot as plt 

epsilon = 0.001
Tid_steps = 30000

#startværdier 
x_start = 1.75
y_start = 0
vx_start = 0.0
vy_start = -1.63/4

#Opstiller arrays til sted, fart, og acceleration
x = [0.0]*Tid_steps
y = [0.0]*Tid_steps
vx = [0.0]*Tid_steps
vy = [0.0]*Tid_steps
ax = [0.0]*Tid_steps
ay = [0.0]*Tid_steps
r = [0.0]*Tid_steps

#opsætter vores bevægelsesligninger 
x[0] = x_start 
y[0] = y_start 

r[0] = math.sqrt(x[0]**2+y[0]**2)
r_3 = r[0]**3
ax[0] = -x[0]/r_3
ay[0] = -y[0]/r_3

vx[0] = vx_start + (epsilon/2.0)*ax[0]
vy[0] = vy_start + (epsilon/2.0)*ay[0]

  
#startværdier for planet 2
x2_start = 1.75
y2_start = 0
vx2_start = 0.0
vy2_start = -1.63/3

#Opstiller arrays til sted, fart, og acceleration for planet 2 
x2 = [0.0]*Tid_steps
y2 = [0.0]*Tid_steps
vx2 = [0.0]*Tid_steps
vy2 = [0.0]*Tid_steps
ax2 = [0.0]*Tid_steps
ay2 = [0.0]*Tid_steps
r2 = [0.0]*Tid_steps

#opsætter vores bevægelsesligninger for planet 2 
x2[0] = x2_start 
y2[0] = y2_start 

r2[0] = math.sqrt(x2[0]**2+y2[0]**2)
r_32 = r2[0]**3
ax2[0] = -x2[0]/r_32
ay2[0] = -y2[0]/r_32

vx2[0] = vx2_start + (epsilon/2.0)*ax2[0]
vy2[0] = vy2_start + (epsilon/2.0)*ay2[0]



#startværdier for planet 3
x3_start = 1.75
y3_start = 0
vx3_start = 0.0
vy3_start = 0.5

#Opstiller arrays til sted, fart, og acceleration for planet 3
x3 = [0.0]*Tid_steps
y3 = [0.0]*Tid_steps
vx3 = [0.0]*Tid_steps
vy3 = [0.0]*Tid_steps
ax3 = [0.0]*Tid_steps
ay3 = [0.0]*Tid_steps
r3 = [0.0]*Tid_steps

#opsætter vores bevægelsesligninger for planet 3 
x3[0] = x3_start 
y3[0] = y3_start 

r3[0] = math.sqrt(x3[0]**2+y3[0]**2)
r_33 = r3[0]**3
ax3[0] = -x3[0]/r_33
ay3[0] = -y3[0]/r_33

vx3[0] = vx3_start + (epsilon/2.0)*ax3[0]
vy3[0] = vy3_start + (epsilon/2.0)*ay3[0]
```

```python
#simuler planet 1 
for i in range(1, Tid_steps):
    x[i] = x[i-1] + epsilon*vx[i-1]
    y[i] = y[i-1] + epsilon*vy[i-1]
    
    r[i] = math.sqrt(x[i]**2+y[i]**2)
    r_3 = r[i]**3
    ax[i] = -x[i]/r_3
    ay[i] = -y[i]/r_3

    vx[i] = vx[i-1] + (epsilon/2.0)*ax[i]
    vy[i] = vy[i-1] + (epsilon/2.0)*ay[i]

#simuler planet 2 
for i in range(1, Tid_steps):
    x2[i] = x2[i-1] + epsilon*vx2[i-1]
    y2[i] = y2[i-1] + epsilon*vy2[i-1]
    
    r2[i] = math.sqrt(x2[i]**2+y2[i]**2)
    r_32 = r2[i]**3
    ax2[i] = -x2[i]/r_32
    ay2[i] = -y2[i]/r_32

    vx2[i] = vx2[i-1] + (epsilon/2.0)*ax2[i]
    vy2[i] = vy2[i-1] + (epsilon/2.0)*ay2[i]
    
#simuler planet 3    
for i in range(1, Tid_steps):
    x3[i] = x3[i-1] + epsilon*vx3[i-1]
    y3[i] = y3[i-1] + epsilon*vy3[i-1]
    
    r3[i] = math.sqrt(x3[i]**2+y3[i]**2)
    r_33 = r3[i]**3
    ax3[i] = -x3[i]/r_33
    ay3[i] = -y3[i]/r_33

    vx3[i] = vx3[i-1] + (epsilon/2.0)*ax3[i]
    vy3[i] = vy3[i-1] + (epsilon/2.0)*ay3[i]

plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.xlabel('x poistion')
plt.ylabel('y position')
plt.grid()
plt.xlim(-2.0,2.0)
plt.ylim(-2.0,2.0)
plt.plot([0], [0], marker='o', markersize=3, color="red")
plt.show
```
